### Write a function varSummary(var) 
which calculates: mean, standard deviation and range for a vector passed as a function argument. The calculated values should be printed to the console as a single line. Values should be separated by space.

```python
def varSummary(var):
    print (var.name)
    #print (var.name)
    # mean
    vmean = numpy.mean(var)
    # standard deviation over requested axis
    vstd = numpy.std(var)
    # range
    vrange = numpy.ptp(var)

    return print ("%.2f %.2f %.2f" % (vmean, vstd, vrange), sep=' ', end='\n')
    # the function will return numbers with precision of 2.
```

### For each continuous variable in the data frame call ''varSummary'' function in a loop. 



```python
m = d.loc[:, 'fixed.acidity':'quality'].apply(varSummary, axis=0)
```

###	Draw a histogram of wine quality.


A histogram is a function mi that counts the number of observations that fall into each of the disjoint categories (known as bins) [ref: https://en.wikipedia.org/wiki/Histogram]

```python
from scipy.stats import kurtosis
doanes = lambda d: int(1 + numpy.log(len(d)) + numpy.log(1 + kurtosis(d) * (len(d) / 6.) ** 0.5))
d.hist(column="quality", grid=False, bins=doanes(d.quality.dropna()))
# d.quality.plot(kind='kde', xlim=(0,600), style='r--')
# number of bins (11) calulcated from Doane's formula, here I used equation from http://nbviewer.jupyter.org/gist/fonnesbeck/5850463 
```
